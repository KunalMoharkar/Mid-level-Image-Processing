```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#utility functions

def threshold_img(img):
    
    ret,img_new = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    return img_new


def erode(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_eroded = np.zeros(img.shape,dtype='uint8')
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            num = np.sum(subMat)
            
            if num == 255*(window_size*window_size):
                im_eroded[x-padAmount][y-padAmount] = 255
    
    
    return im_eroded

def dilate(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_dilated = np.zeros(img.shape,dtype='uint8')
    
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            
            subMat = np.reshape(subMat,-1)
            
            if 255 in subMat:
                im_dilated[x-padAmount][y-padAmount] = 255
    
    
    return im_dilated



```

```python
img = cv2.imread("images/fingerprint.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print(img.shape)

img = threshold_img(img)


plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])


img_1 = erode(img,5)

print(img_1.sum())

plt.subplot(2,2,2)
plt.imshow(img_1,cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])


img_1 = dilate(img_1,5)

img_1 = dilate(img_1,5)

img_1 = erode(img_1,5)

plt.subplot(2,2,3)
plt.imshow(img_1,cmap='gray')
plt.title('Dilated Image')
plt.xticks([])
plt.yticks([])
```

```python
def is_neighbour(x,y,img):
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    if x-1 >= 0 and y-1 >=0 and x+1 < ht and y+1 <wt and img[x][y] == 0:
        return True
        
    return False

def count_connected(img):
    
    vis = np.zeros(img.shape, dtype='uint')
      
    count = 0
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    queue = []
    
    for x in range(ht):   
    
        for y in range(wt):
            
            if vis[x][y] == 0 and img[x][y]==0:
                
            
                queue.append((x,y))
                vis[x][y] == 1
                
                count = count + 1
                
                
                while(len(queue) != 0):
                    
                    
                    a, b = queue.pop(0)
                    
                    
                    if is_neighbour(a+1,b,img) and vis[a+1][b] == 0:
                        queue.append((a+1,b))
                        vis[a+1][b] = 1
                    
                    if is_neighbour(a+1,b+1,img) and vis[a+1][b+1] == 0:
                        queue.append((a+1,b+1))
                        vis[a+1][b+1] = 1
                    
                    if is_neighbour(a+1,b-1,img) and vis[a+1][b-1] == 0:
                        queue.append((a+1,b-1))
                        vis[a+1][b-1] = 1
                    
                    if is_neighbour(a,b+1,img) and vis[a][b+1] == 0:
                        queue.append((a,b+1))
                        vis[a][b+1] = 1
                    
                    if is_neighbour(a,b-1,img) and vis[a][b-1] == 0:
                        queue.append((a,b-1))
                        vis[a][b-1] = 1
                    
                    if is_neighbour(a-1,b+1,img) and vis[a-1][b+1] == 0:
                        queue.append((a-1,b+1))
                        vis[a-1][b+1] = 1
                        
                    if is_neighbour(a-1,b-1,img) and vis[a-1][b-1] == 0:
                        queue.append((a-1,b-1))
                        vis[a-1][b-1] = 1
                    
                    if is_neighbour(a-1,b,img) and vis[a-1][b] == 0:
                        queue.append((a-1,b))
                        vis[a-1][b] = 1
                    
                
    return count
            



img = cv2.imread("images/disks.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

img = threshold_img(img)

print(img)

print(img.shape)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])


img = dilate(img,41)

print(count_connected(img))

plt.subplot(2,2,2)
plt.imshow(img,cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])
```

```python

def Myrgb2hsi(img):
    
    #normailze
    img = np.float32(img)/255
    
    b = img[:,:,0]
    g = img[:,:,1]
    r = img[:,:,2]
    
    
    #intensity component calucation
    
    img_intensity = np.divide(b + g + r, 3)
    
    plt.figure(figsize=(16,12))
    plt.subplot(2,2,1)
    plt.imshow(img_intensity,cmap='gray')
    plt.title('Intensity Image')
    plt.xticks([])
    plt.yticks([])
    
    #staturation component calculation
    
    minimum = np.minimum(np.minimum(r, g), b)
    img_saturation = 1 - (3 / (r + g + b) * minimum)
    
    plt.subplot(2,2,2)
    plt.imshow(img_saturation,cmap='gray')
    plt.title('Saturation Image')
    plt.xticks([])
    plt.yticks([])
    
    
    #hue componenet calculation
    
    hue = r.copy()

    for i in range(0, b.shape[0]):
        for j in range(0, b.shape[1]):
            hue[i][j] = 0.5 * ((r[i][j] - g[i][j]) + (r[i][j] - b[i][j])) / \
                                math.sqrt((r[i][j] - g[i][j])**2 +
                                        ((r[i][j] - b[i][j]) * (g[i][j] - b[i][j])) + 0.00000000001)
            hue[i][j] = math.acos(hue[i][j])

            if b[i][j] <= g[i][j]:
                hue[i][j] = hue[i][j]
            else:
                hue[i][j] = ((360 * math.pi) / 180.0) - hue[i][j]
    
    plt.subplot(2,2,3)
    plt.imshow(hue,cmap='gray')
    plt.title('Hue Image')
    plt.xticks([])
    plt.yticks([])
    
img = cv2.imread("images/ball.bmp")

img_hsi = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

im_h = img_hsi[:,:,0]
im_s = img_hsi[:,:,1]
im_i = img_hsi[:,:,2]


plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(im_h, cmap='gray')
plt.title('inbilt Image hue')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(im_s, cmap='gray')
plt.title('inbilt Image saturation')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(im_i, cmap='gray')
plt.title('inbilt Image intensity')
plt.xticks([])
plt.yticks([])


Myrgb2hsi(img)

cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

print(img.shape)


```

```python
def findCorners(img, window_size, k, thresh):
    dy, dx = np.gradient(img)
    Ixx = dx**2
    Ixy = dy*dx
    Iyy = dy**2
    height = img.shape[0]
    width = img.shape[1]
    
    print((height,width))

    cornerList = []
    newImg = img.copy()
    offset = int(window_size/2)


    for y in range(offset, height-offset):
        for x in range(offset, width-offset):
            #Calculate sum of squares
            windowIxx = Ixx[y-offset:y+offset+1, x-offset:x+offset+1]
            windowIxy = Ixy[y-offset:y+offset+1, x-offset:x+offset+1]
            windowIyy = Iyy[y-offset:y+offset+1, x-offset:x+offset+1]
            Sxx = windowIxx.sum()
            Sxy = windowIxy.sum()
            Syy = windowIyy.sum()


            det = (Sxx * Syy) - (Sxy**2)
            trace = Sxx + Syy
            r = det - k*(trace**2)
        

            if r > thresh:
                
                cornerList.append((y,x))

                #color_img.itemset((y, x), 255)
                
    return cornerList



img = cv2.imread("images/corner_test.png")

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)

img[dst > 0.01 * dst.max()]=[0, 0, 255]

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.title('Original img')
plt.xticks([])
plt.yticks([])



dst = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(2,2,2)
plt.imshow(dst)
plt.title('corner identified')
plt.xticks([])
plt.yticks([])


cornerList = findCorners(gray,3,0.04,30000)

for px in cornerList:
    
    img[px[0],px[1],0] = 0
    img[px[0],px[1],1] = 0
    img[px[0],px[1],2] = 255



img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(2,2,3)
plt.imshow(img)
plt.title('corner identified')
plt.xticks([])
plt.yticks([])

```
