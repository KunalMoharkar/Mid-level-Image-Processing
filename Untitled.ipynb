```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#utility functions

def threshold_img(img):
    
    ret,img_new = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    return img_new


def erode(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_eroded = np.zeros(img.shape,dtype='uint8')
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            num = np.sum(subMat)
            
            if num == 255*(window_size*window_size):
                im_eroded[x-padAmount][y-padAmount] = 255
    
    
    return im_eroded

def dilate(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_dilated = np.zeros(img.shape,dtype='uint8')
    
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            
            subMat = np.reshape(subMat,-1)
            
            if 255 in subMat:
                im_dilated[x-padAmount][y-padAmount] = 255
    
    
    return im_dilated



```

```python
img = cv2.imread("images/fingerprint.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print(img.shape)

img = threshold_img(img)


plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])


img_1 = erode(img,5)

print(img_1.sum())

plt.subplot(2,2,2)
plt.imshow(img_1,cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])


img_1 = dilate(img_1,5)

img_1 = dilate(img_1,5)

img_1 = erode(img_1,5)

plt.subplot(2,2,3)
plt.imshow(img_1,cmap='gray')
plt.title('Dilated Image')
plt.xticks([])
plt.yticks([])
```

```python
def is_neighbour(x,y,img):
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    if x-1 >= 0 and y-1 >=0 and x+1 < ht and y+1 <wt and img[x][y] == 0:
        return True
        
    return False






def count_connected(img):
    
    vis = np.zeros(img.shape, dtype='uint')
    
    
    count = 0
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    queue = []
    
    for x in range(ht):
        
    
        for y in range(wt):
            
            if vis[x][y] == 0 and img[x][y]==0:
                
            
                queue.append((x,y))
                vis[x][y] == 1
                
                count = count + 1
                
                
                while(len(queue) != 0):
                    
                    
                    a, b = queue.pop(0)
                    
                    
                    if is_neighbour(a+1,b,img) and vis[a+1][b] == 0:
                        queue.append((a+1,b))
                        vis[a+1][b] = 1
                    
                    if is_neighbour(a+1,b+1,img) and vis[a+1][b+1] == 0:
                        queue.append((a+1,b+1))
                        vis[a+1][b+1] = 1
                    
                    if is_neighbour(a+1,b-1,img) and vis[a+1][b-1] == 0:
                        queue.append((a+1,b-1))
                        vis[a+1][b-1] = 1
                    
                    if is_neighbour(a,b+1,img) and vis[a][b+1] == 0:
                        queue.append((a,b+1))
                        vis[a][b+1] = 1
                    
                    if is_neighbour(a,b-1,img) and vis[a][b-1] == 0:
                        queue.append((a,b-1))
                        vis[a][b-1] = 1
                    
                    if is_neighbour(a-1,b+1,img) and vis[a-1][b+1] == 0:
                        queue.append((a-1,b+1))
                        vis[a-1][b+1] = 1
                        
                    if is_neighbour(a-1,b-1,img) and vis[a-1][b-1] == 0:
                        queue.append((a-1,b-1))
                        vis[a-1][b-1] = 1
                    
                    if is_neighbour(a-1,b,img) and vis[a-1][b] == 0:
                        queue.append((a-1,b))
                        vis[a-1][b] = 1
                    
                
                    
        
        
        
        
        
        
        
        
        
                
    return count
            





img = cv2.imread("images/disks.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

img = threshold_img(img)

print(img)

print(img.shape)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])


img = dilate(img,41)

print(count_connected(img))

plt.subplot(2,2,2)
plt.imshow(img,cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])
```
