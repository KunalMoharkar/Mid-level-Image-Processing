```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#utility functions

def threshold_img(img):
    
    ret,img_new = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    return img_new


def erode(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_eroded = np.zeros(img.shape,dtype='uint8')
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            num = np.sum(subMat)
            
            if num == 255*(window_size*window_size):
                im_eroded[x-padAmount][y-padAmount] = 255
    
    
    return im_eroded

def dilate(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_dilated = np.zeros(img.shape,dtype='uint8')
    
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            
            subMat = np.reshape(subMat,-1)
            
            if 255 in subMat:
                im_dilated[x-padAmount][y-padAmount] = 255
    
    
    return im_dilated



```

```python
#read image
img = cv2.imread("images/fingerprint.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#binary threshld image
img = threshold_img(img)


#original image
plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

#opening operation
img_1 = erode(img,5)
img_1 = dilate(img_1,5)

plt.subplot(2,2,2)
plt.imshow(img_1,cmap='gray')
plt.title('Opening Image')
plt.xticks([])
plt.yticks([])



#closing operation
img_1 = dilate(img_1,5)
img_1 = erode(img_1,5)

plt.subplot(2,2,3)
plt.imshow(img_1,cmap='gray')
plt.title('Final Image')
plt.xticks([])
plt.yticks([])
```

```python
#check if a neighbour is valid 
def is_neighbour(x,y,img):
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    if x-1 >= 0 and y-1 >=0 and x+1 < ht and y+1 <wt and img[x][y] == 0:
        return True
        
    return False



#function to count discs
def count_connected(img):
    
    vis = np.zeros(img.shape, dtype='uint')
      
    count = 0
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    queue = []
    
    #BFS traversal
    for x in range(ht):   
    
        for y in range(wt):
            
            if vis[x][y] == 0 and img[x][y]==0:
                
            
                queue.append((x,y))
                vis[x][y] == 1
                
                count = count + 1
            
                
                while(len(queue) != 0):
                    
                    
                    a, b = queue.pop(0)
                    
                    
                    if is_neighbour(a+1,b,img) and vis[a+1][b] == 0:
                        queue.append((a+1,b))
                        vis[a+1][b] = 1
                    
                    if is_neighbour(a+1,b+1,img) and vis[a+1][b+1] == 0:
                        queue.append((a+1,b+1))
                        vis[a+1][b+1] = 1
                    
                    if is_neighbour(a+1,b-1,img) and vis[a+1][b-1] == 0:
                        queue.append((a+1,b-1))
                        vis[a+1][b-1] = 1
                    
                    if is_neighbour(a,b+1,img) and vis[a][b+1] == 0:
                        queue.append((a,b+1))
                        vis[a][b+1] = 1
                    
                    if is_neighbour(a,b-1,img) and vis[a][b-1] == 0:
                        queue.append((a,b-1))
                        vis[a][b-1] = 1
                    
                    if is_neighbour(a-1,b+1,img) and vis[a-1][b+1] == 0:
                        queue.append((a-1,b+1))
                        vis[a-1][b+1] = 1
                        
                    if is_neighbour(a-1,b-1,img) and vis[a-1][b-1] == 0:
                        queue.append((a-1,b-1))
                        vis[a-1][b-1] = 1
                    
                    if is_neighbour(a-1,b,img) and vis[a-1][b] == 0:
                        queue.append((a-1,b))
                        vis[a-1][b] = 1
                    
                
    return count
            


#read image
img = cv2.imread("images/disks.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#threshold
img = threshold_img(img)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

#erode to disconnect discs
#note we call dialte as disks are black here 
#0 value pixels
img = dilate(img,41)

plt.subplot(2,2,2)
plt.imshow(img,cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])


print(f"Total Discs are: {count_connected(img)}")

```

```python

def Myrgb2hsi(img):
    
    #normailze
    img = np.float32(img)/255
    
    b = img[:,:,0]
    g = img[:,:,1]
    r = img[:,:,2]
    
    #intensity component calucation
    img_intensity = np.divide(b + g + r, 3)
    
    plt.figure(figsize=(16,12))
    plt.subplot(2,2,1)
    plt.imshow(img_intensity,cmap='gray')
    plt.title('Intensity Image')
    plt.xticks([])
    plt.yticks([])
    
    #staturation component calculation
    
    minimum = np.minimum(np.minimum(r, g), b)
    img_saturation = 1 - (3 / (r + g + b) * minimum)
    
    plt.subplot(2,2,2)
    plt.imshow(img_saturation,cmap='gray')
    plt.title('Saturation Image')
    plt.xticks([])
    plt.yticks([])
    
    
    #hue componenet calculation
    
    hue = r.copy()

    for i in range(0, b.shape[0]):
        for j in range(0, b.shape[1]):
            hue[i][j] = 0.5 * ((r[i][j] - g[i][j]) + (r[i][j] - b[i][j])) / \
                                math.sqrt((r[i][j] - g[i][j])**2 +
                                        ((r[i][j] - b[i][j]) * (g[i][j] - b[i][j])) + 0.00000000001)
            
            hue[i][j] = math.acos(hue[i][j])
            

            if b[i][j] <= g[i][j]:
                hue[i][j] = hue[i][j]
            else:
                hue[i][j] = ((360 * math.pi) / 180.0) - hue[i][j]
    
    plt.subplot(2,2,3)
    plt.imshow(hue,cmap='gray')
    plt.title('Hue Image')
    plt.xticks([])
    plt.yticks([])
    
    return img_intensity,img_saturation,hue
    
    
img = cv2.imread("images/ball.bmp")

img_hsi = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

#normailze
img_hsi = np.float32(img_hsi)/255

im_h = img_hsi[:,:,0]
im_s = img_hsi[:,:,1]
im_i = img_hsi[:,:,2]

print(im_h)


cv2.imshow('hue',im_h)
cv2.waitKey(0)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(im_h, cmap='gray')
plt.title('inbilt Image hue')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(im_s, cmap='gray')
plt.title('inbilt Image saturation')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(im_i, cmap='gray')
plt.title('inbilt Image intensity')
plt.xticks([])
plt.yticks([])


imh_i,imh_s,imh_h = Myrgb2hsi(img)

diff_i = imh_i - im_i
diff_s = imh_s - im_s
diff_h = imh_h - im_h

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(diff_h, cmap='gray')
plt.title('Diff Image hue')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(diff_s, cmap='gray')
plt.title('Diff Image saturation')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(diff_i, cmap='gray')
plt.title('Diff Image intensity')
plt.xticks([])
plt.yticks([])




cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

print(img.shape)


```

```python
def findCorners(img, window_size, k, thresh):
    
    #calucate image gradient
    dy, dx = np.gradient(img)
    Ixx = dx**2
    Ixy = dy*dx
    Iyy = dy**2
    
    height = img.shape[0]
    width = img.shape[1]
    

    cornerList = []
    newImg = img.copy()
    offset = int(window_size/2)

    for y in range(offset, height-offset):
        for x in range(offset, width-offset):
        
            windowIxx = Ixx[y-offset:y+offset+1, x-offset:x+offset+1]
            windowIxy = Ixy[y-offset:y+offset+1, x-offset:x+offset+1]
            windowIyy = Iyy[y-offset:y+offset+1, x-offset:x+offset+1]
            
            #for each neighbourung window calucate summatioon
            Sxx = windowIxx.sum()
            Sxy = windowIxy.sum()
            Syy = windowIyy.sum()


            det = (Sxx * Syy) - (Sxy**2)
            trace = Sxx + Syy
            
            #calcuate the response
            r = det - k*(trace**2)
        
            
            #appen to list only if response greater the the threshold
            if r > thresh:
                
                cornerList.append((y,x))
                
    return cornerList



img = cv2.imread("images/corner_test.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

#call inbuilt harris corner function
dst = cv2.cornerHarris(gray,2,3,0.04)

#plot the response points 
#choose 0.01*max response as threshold
img_new = img.copy()
img_new[dst > 0.01 * dst.max()]=[255,0 , 0]

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.title('Original img')
plt.xticks([])
plt.yticks([])


img_new = cv2.cvtColor(img_new,cv2.COLOR_BGR2RGB)

plt.subplot(2,2,2)
plt.imshow(img_new)
plt.title('corner identified')
plt.xticks([])
plt.yticks([])

#harndcoded harris corner
cornerList = findCorners(gray,3,0.04,0.01 * dst.max())

for px in cornerList:
    
    img[px[0],px[1],0] = 255
    img[px[0],px[1],1] = 0
    img[px[0],px[1],2] = 0



img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(2,2,3)
plt.imshow(img)
plt.title('corner identified handcoded')
plt.xticks([])
plt.yticks([])

```
