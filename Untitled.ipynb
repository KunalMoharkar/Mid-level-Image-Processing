```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


img = cv2.imread("images/fingerprint.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print(img)

ret,img1 = cv2.threshold(img,254,255,cv2.THRESH_BINARY)

print(img1)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Mask Image')
plt.xticks([])
plt.yticks([])
plt.subplot(2,2,2)
plt.imshow(img1,cmap='gray')
plt.title('Inverse Mask Image')
plt.xticks([])
plt.yticks([])
```
