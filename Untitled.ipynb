```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#utility functions

def threshold_img(img):
    
    ret,img_new = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    return img_new


def erode(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_eroded = np.zeros(img.shape,dtype='uint8')
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            num = np.sum(subMat)
            
            if num == 255*(window_size*window_size):
                im_eroded[x-padAmount][y-padAmount] = 255
    
    
    return im_eroded

def dilate(img,window_size):
    
    padAmount = int(window_size/2)
    
    im_dilated = np.zeros(img.shape,dtype='uint8')
    
    #add a zero padding
    img= cv2.copyMakeBorder(img,padAmount,padAmount,padAmount,padAmount,cv2.BORDER_CONSTANT,value=0)
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    colstart = padAmount
    rowstart = padAmount
    colend = wt - padAmount
    rowend = ht - padAmount
    
    for x in range(rowstart,rowend):
        
        for y in range(colstart,colend):
            
            subMat = img[x-padAmount:x+padAmount+1,y-padAmount:y+padAmount+1]
            
            subMat = np.reshape(subMat,-1)
            
            if 255 in subMat:
                im_dilated[x-padAmount][y-padAmount] = 255
    
    
    return im_dilated



```

```python
#read image
img = cv2.imread("images/fingerprint.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#binary threshld image
img = threshold_img(img)


#original image
plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

#opening operation
img_1 = erode(img,5)
img_1 = dilate(img_1,5)

plt.subplot(2,2,2)
plt.imshow(img_1,cmap='gray')
plt.title('Opening Image')
plt.xticks([])
plt.yticks([])



#closing operation
img_1 = dilate(img_1,5)
img_1 = erode(img_1,5)

plt.subplot(2,2,3)
plt.imshow(img_1,cmap='gray')
plt.title('Final Image')
plt.xticks([])
plt.yticks([])
```

```python
#check if a neighbour is valid 
def is_neighbour(x,y,img):
    
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    if x-1 >= 0 and y-1 >=0 and x+1 < ht and y+1 <wt and img[x][y] == 0:
        return True
        
    return False



#function to count discs
def count_connected(img):
    
    vis = np.zeros(img.shape, dtype='uint')
      
    count = 0
    
    ht = img.shape[0]
    wt = img.shape[1]
    
    queue = []
    
    #BFS traversal
    for x in range(ht):   
    
        for y in range(wt):
            
            if vis[x][y] == 0 and img[x][y]==0:
                
            
                queue.append((x,y))
                vis[x][y] == 1
                
                count = count + 1
            
                
                while(len(queue) != 0):
                    
                    
                    a, b = queue.pop(0)
                    
                    
                    if is_neighbour(a+1,b,img) and vis[a+1][b] == 0:
                        queue.append((a+1,b))
                        vis[a+1][b] = 1
                    
                    if is_neighbour(a+1,b+1,img) and vis[a+1][b+1] == 0:
                        queue.append((a+1,b+1))
                        vis[a+1][b+1] = 1
                    
                    if is_neighbour(a+1,b-1,img) and vis[a+1][b-1] == 0:
                        queue.append((a+1,b-1))
                        vis[a+1][b-1] = 1
                    
                    if is_neighbour(a,b+1,img) and vis[a][b+1] == 0:
                        queue.append((a,b+1))
                        vis[a][b+1] = 1
                    
                    if is_neighbour(a,b-1,img) and vis[a][b-1] == 0:
                        queue.append((a,b-1))
                        vis[a][b-1] = 1
                    
                    if is_neighbour(a-1,b+1,img) and vis[a-1][b+1] == 0:
                        queue.append((a-1,b+1))
                        vis[a-1][b+1] = 1
                        
                    if is_neighbour(a-1,b-1,img) and vis[a-1][b-1] == 0:
                        queue.append((a-1,b-1))
                        vis[a-1][b-1] = 1
                    
                    if is_neighbour(a-1,b,img) and vis[a-1][b] == 0:
                        queue.append((a-1,b))
                        vis[a-1][b] = 1
                    
                
    return count
            


#read image
img = cv2.imread("images/disks.png")
img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#threshold
img = threshold_img(img)

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

#erode to disconnect discs
#note we call dialte as disks are black here 
#0 value pixels
img = dilate(img,41)

plt.subplot(2,2,2)
plt.imshow(img,cmap='gray')
plt.title('Eroded Image')
plt.xticks([])
plt.yticks([])


print(f"Total Discs are: {count_connected(img)}")

```

```python

def Myrgb2hsi(img):
    
    #normailze
    img = np.float32(img)/255
    
    b = img[:,:,0]
    g = img[:,:,1]
    r = img[:,:,2]
    
    #intensity component calucation
    img_intensity = np.divide(b + g + r, 3)
    
   
    
    #staturation component calculation
    
    minimum = np.minimum(np.minimum(r, g), b)
    img_saturation = 1 - (3 / (r + g + b) * minimum)
    
    
    
    
    #hue componenet calculation
    
    hue = r.copy()

    for i in range(0, b.shape[0]):
        for j in range(0, b.shape[1]):
            hue[i][j] = 0.5 * ((r[i][j] - g[i][j]) + (r[i][j] - b[i][j])) / \
                                math.sqrt((r[i][j] - g[i][j])**2 +
                                        ((r[i][j] - b[i][j]) * (g[i][j] - b[i][j])) + 0.00000000001)
            
            hue[i][j] = math.acos(hue[i][j])
            

            if b[i][j] <= g[i][j]:
                hue[i][j] = hue[i][j]
            else:
                hue[i][j] = ((360 * math.pi) / 180.0) - hue[i][j]
    
    
    return img_intensity,img_saturation,hue
    
    
img = cv2.imread("images/ball.bmp")

img_hsi = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

#normailze

img_hsi = np.float32(img_hsi)/255

im_h = img_hsi[:,:,0]
im_s = img_hsi[:,:,1]
im_i = img_hsi[:,:,2]

#plot inbuilt figures

plt.figure(figsize=(16,12))
plt.subplot(1,3,1)
plt.imshow(im_h, cmap='gray')
plt.title('inbilt Image hue')
plt.xticks([])
plt.yticks([])

plt.subplot(1,3,2)
plt.imshow(im_s, cmap='gray')
plt.title('inbilt Image saturation')
plt.xticks([])
plt.yticks([])

plt.subplot(1,3,3)
plt.imshow(im_i, cmap='gray')
plt.title('inbilt Image intensity')
plt.xticks([])
plt.yticks([])


#hancoded functio call
imh_i,imh_s,imh_h = Myrgb2hsi(img)




#plot hnadcoded figure
plt.figure(figsize=(16,12))
plt.subplot(1,3,1)
plt.imshow(imh_h,cmap='gray')
plt.title('Hue Image')
plt.xticks([])
plt.yticks([])


plt.subplot(1,3,2)
plt.imshow(imh_s,cmap='gray')
plt.title('Saturation Image')
plt.xticks([])
plt.yticks([])


plt.subplot(1,3,3)
plt.imshow(imh_i,cmap='gray')
plt.title('intensity Image')
plt.xticks([])
plt.yticks([])


#plot the differenec figures

diff_i = imh_i - im_i
diff_s = imh_s - im_s
diff_h = imh_h - im_h

plt.figure(figsize=(16,12))
plt.subplot(1,3,1)
plt.imshow(diff_h, cmap='gray')
plt.title('Diff Image hue')
plt.xticks([])
plt.yticks([])

plt.subplot(1,3,2)
plt.imshow(diff_s, cmap='gray')
plt.title('Diff Image saturation')
plt.xticks([])
plt.yticks([])

plt.subplot(1,3,3)
plt.imshow(diff_i, cmap='gray')
plt.title('Diff Image intensity')
plt.xticks([])
plt.yticks([])


#thresholding in hue space

imh_h = imh_h/np.max(imh_h)
imh_h = imh_h*255
imh_h = imh_h.astype('uint8')
ret,imh_h = cv2.threshold(imh_h,30,255,cv2.THRESH_BINARY_INV)

plt.figure(figsize=(16,12))
plt.subplot(1,3,1)
plt.imshow(imh_h, cmap='gray')
plt.title('Threshold Image hue Space')
plt.xticks([])
plt.yticks([])


#morphological processing
ele = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(7,7))
imh_h = cv2.morphologyEx(imh_h, cv2.MORPH_OPEN, ele)
imh_h = cv2.morphologyEx(imh_h, cv2.MORPH_CLOSE, ele)


#centroid calculation

M = cv2.moments(imh_h)
cX = int(M["m10"] / M["m00"])
cY = int(M["m01"] / M["m00"])

#mark centroid on original image woth green color
cv2.circle(img, (cX, cY), 5, (0,255,20), -1)

plt.subplot(1,2,1)
plt.imshow(imh_h, cmap='gray')
plt.title('Threshold Image hue Space')
plt.xticks([])
plt.yticks([])

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.subplot(1,2,1)
plt.imshow(img)
plt.title('marked image Space')
plt.xticks([])
plt.yticks([])




```

```python
def findCorners(img, window_size, k, thresh):
    
    #calucate image gradient
    dy, dx = np.gradient(img)
    Ixx = dx**2
    Ixy = dy*dx
    Iyy = dy**2
    
    height = img.shape[0]
    width = img.shape[1]
    

    cornerList = []
    newImg = img.copy()
    offset = int(window_size/2)

    for y in range(offset, height-offset):
        for x in range(offset, width-offset):
        
            windowIxx = Ixx[y-offset:y+offset+1, x-offset:x+offset+1]
            windowIxy = Ixy[y-offset:y+offset+1, x-offset:x+offset+1]
            windowIyy = Iyy[y-offset:y+offset+1, x-offset:x+offset+1]
            
            #for each neighbourung window calucate summatioon
            Sxx = windowIxx.sum()
            Sxy = windowIxy.sum()
            Syy = windowIyy.sum()


            det = (Sxx * Syy) - (Sxy**2)
            trace = Sxx + Syy
            
            #calcuate the response
            r = det - k*(trace**2)
        
            
            #appen to list only if response greater the the threshold
            if r > thresh:
                
                cornerList.append((y,x))
                
    return cornerList



img = cv2.imread("images/corner_test.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

#call inbuilt harris corner function
dst = cv2.cornerHarris(gray,2,3,0.04)

#plot the response points 
#choose 0.01*max response as threshold
img_new = img.copy()
img_new[dst > 0.01 * dst.max()]=[255,0 , 0]

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.title('Original img')
plt.xticks([])
plt.yticks([])


img_new = cv2.cvtColor(img_new,cv2.COLOR_BGR2RGB)

plt.subplot(2,2,2)
plt.imshow(img_new)
plt.title('corner identified')
plt.xticks([])
plt.yticks([])

#harndcoded harris corner
cornerList = findCorners(gray,3,0.04,0.01 * dst.max())

for px in cornerList:
    
    img[px[0],px[1],0] = 255
    img[px[0],px[1],1] = 0
    img[px[0],px[1],2] = 0



img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(2,2,3)
plt.imshow(img)
plt.title('corner identified handcoded')
plt.xticks([])
plt.yticks([])




```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

MIN_MATCH_COUNT = 10

img1 = cv2.imread('images/FM_img1.jpg',0)          # queryImage
img2 = cv2.imread('images/FM_img2.jpg',0) # trainImage

print(img1)
print(img2)

sift = cv2.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)

FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)

flann = cv2.FlannBasedMatcher(index_params, search_params)

matches = flann.knnMatch(des1,des2,k=2)

# store all the good matches as per Lowe's ratio test.
good = []
for m,n in matches:
    if m.distance < 0.001*n.distance:
        good.append(m)
        
    src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
    dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)

    M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
    matchesMask = mask.ravel().tolist()

    h,w = img1.shape
    pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    #dst = cv2.perspectiveTransform(pts,M)

    img2 = cv2.polylines(img2,[np.int32(pts)],True,255,3, cv2.LINE_AA)

else:
   # print "Not enough matches are found - %d/%d" % (len(good),MIN_MATCH_COUNT)
    matchesMask = None
    
draw_params = dict(matchColor = (0,255,0), # draw matches in green color
                   singlePointColor = None,
                   matchesMask = matchesMask, # draw only inliers
                   flags = 2)

img3 = cv2.drawMatches(img1,kp1,img2,kp2,good,None,**draw_params)

plt.imshow(img3, 'gray'),plt.show()
```
